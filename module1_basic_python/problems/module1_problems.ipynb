#### 建立資料夾 support.py -> mkdir(dirpath)

使用方式：mkdir(新資料夾路徑）

這只是一個很簡單的製作資料夾的函數，因此你必須確認你的路徑上的資料夾都是存在的。

這是什麼意思呢？假設你的資料夾有兩個資料夾

```
data
new_data
```

這時你如果想要在 new_data 裡面新增資料夾 花蓮縣，你必須下的指令為

```
mkdir('new_data/花蓮縣')
```

但如果你想在 new_data 裡面新增 臺北市，然後在臺北市裡面新增 大安區，你不能這樣操作

```
mkdir('臺北市/大安區')
```

你必須這樣操作

```
mkdir('臺北市')
mkdir('臺北市/大安區')
```

#### 複製檔案 support.py -> copy()

使用方式：copy('from.txt', 'to.txt')

```
copy('data/3_709001300000000_2016-01-16.txt', 'new_data/花蓮縣/宋楚瑜_花蓮縣卓溪鄉_2016-01-16.txt')
copy('data/2_711000400000000_2016-01-16.txt', 'new_data/澎湖縣/蔡英文_澎湖縣西嶼鄉_2016-01-16.txt')
```

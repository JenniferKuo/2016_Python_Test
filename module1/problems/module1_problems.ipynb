## 解決思路

因為這是第一個問題，這邊有給對於解決問題不太熟悉的同學的一些提示，但這只是個人的一些感想，它不是你唯一解決問題的思考方法，你可以用自己習慣熟悉的方式進行思考。

### Step 1.「來自哪裡，去向何方」

通常我們想要使用程式解決一個問題時，一定要仔細的審題，先搞清楚我們有什麼東西？要做的是什麼？在這個情境中我們可以很清楚地看到，我們有的是「一堆名字散亂的檔案」，而我們想要得到的是「一堆整理好的檔案」。

### Step 2. 將問題分解成小問題

在釐清了我們到底想要得到什麼後，我們得將一個大問題切割成小問題，將解決問題的過程分割成一個一個的小步驟。

以上面的情境來說，我們可能會需要解決「讀入一堆檔案」、「幫很多檔案命名」、「更改日期」、「對應到某些姓名或地名」和「輸出結果」這些小問題，當然問題要分割到多小或怎麼樣去劃界線，這個就跟個人的思考方式有關了。

將問題的複雜程度降低後，我們便能夠針對不同的問題設計不同的程式，一旦我們能夠將所有的小問題解決，自然而然就能夠解決一個龐大的問題（寫很多個簡單的小程式會比寫一個巨大的程式來得容易吧？想想看為什麼？）。

這樣的做法在概念上又被稱為 divide and conquer（演算法裡面也有這樣的東西，概念上是相通的）。

### Step 3. 單元測試

這是被很多人遺忘的部分，為了程式的品質，在真實世界的大型的系統中，每個小程式都會被要求要完整的測試，是因為當程式愈來愈多時，一旦有個地方出錯會變得非常難以解決。不過我們現在的目標只是「用程式解決問題」，這個之後再說吧。

分割完問題後，我們就得將各式小問題的解法，串接成一個完整的程式，這個部分（可能）才是程序猿最重要的工作，也是最需要發揮你們巧思跟所學知識的地方。~~所以程序猿就被當成了很會複製貼上的猴子。~~

### Hint：工作目錄的概念

當我們執行程式時，會有一個 **工作目錄** 的概念，可以簡單想像成「目前這支程式執行的所在資料夾」。

如果你有檔案目錄的概念就會知道，如果我今天想要存取 exercise1/data 這一個資料夾，我可以給予絕對路徑（在Windows 下就是從磁碟機編號開始寫，例如 C:\\exercise1\\data），或者是相對路徑（假設我打開 exercise1 這一個資料夾，我只要點一下 data 就可以打開 exercise/data，這時在系統中可以用 ./data 來表達，. 表示目前這個資料夾），在程式當中如果你需要存取到系統中的其他檔案也同樣適用這個概念。

如果你使用 Sublime Text 2 去運行 exercise1 底下的上面的測試 script，你基本上不會發現任何問題，然而你如果使用指令介面來運行 Python script，可能就會遇到問題了，Python 可能會跟你說他找不到 'data' 這一個資料夾，原因是「你執行 python answer.py 這個指令時，你所在的工作目錄不對」。你可以先使用 cd 移動到 exercise1 這個資料夾後執行你的程式（假設是 answer.py），再到 problems 資料夾執行 exercise1/answer.py，看看輸出結果有什麼不一樣。也許就能明白這個概念。

由於這堂課不是計算機概論，因此只是淺淺帶過一下，如果你對於這個概念感到很模糊，可以投 Q&A，再詳細解釋給你。
